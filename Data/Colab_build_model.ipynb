```python
def translate_and_append(item):
    try:
        translation = translator.translate(item, dest="vi")
        trans.append(translation.text)
        language.append("vi")
    except Exception as e:
        # Retry translation with a sleep (avoid API quota limitations)
        time.sleep(1)
        try:
            translation = translator.translate(item, dest="vi")
            trans.append(translation.text)
            language.append("vi")
        except Exception as e:
            # Try translating to English
            try:
                translation = translator.translate(item, dest="en")
                trans.append(translation.text)
                language.append("en")
            except Exception as e:
                # Use original text as fallback and mark as Portuguese
                trans.append(item)
                language.append("pt")
                print(f"Error translating: {e}")

trans = []
language = []

# Loop through each item in the 'text' column and translate it
for item in df['text']:
    translate_and_append(item)
```

```python
# IMPORT THƯ VIỆN
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, Dataset
from torch import nn
from tqdm import tqdm
import torch
from sklearn.metrics import accuracy_score

# Map sentiment strings to integers
sentiment_mapping = {
    'Positive': 2,
    'Neutral': 1,
    'Negative': 0
}

# Đọc dữ liệu Dataframe
df = pd.read_excel('/content/drive/MyDrive/AI/dataKFC_Final.xlsx')
df['Sentiment'] = df['Sentiment'].map(sentiment_mapping)

# Split the data into train and test sets Chia dữ liệu TRAINING:TESTING = 80:20
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Create a PyTorch Dataset Đưa dữ liệu về dạng tensor
class CustomDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        text = self.dataframe.iloc[idx]['Review']
        label = torch.tensor(self.dataframe.iloc[idx]['Sentiment'], dtype=torch.long)
        encoding = self.tokenizer(text, truncation=True, padding='max_length', max_length=self.max_length, return_tensors='pt')
        return {'input_ids': encoding['input_ids'].flatten(), 'attention_mask': encoding['attention_mask'].flatten(), 'labels': label}
# CustomDataset: Lớp Dataset tùy chỉnh của PyTorch để xử lý dữ liệu văn bản và chuyển đổi nó thành tensors.
# __init__: Khởi tạo lớp với dataframe, tokenizer và max_length.
# __len__: Trả về số lượng mẫu trong dataframe.
# __getitem__: Lấy một mẫu từ dataframe tại chỉ số idx, chuyển đổi văn bản thành tensor và trả về dictionary chứa input_ids,
#attention_mask và label.

# Initialize the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
num_labels = len(df['Sentiment'])
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)
# BertTokenizer.from_pretrained: Khởi tạo tokenizer từ mô hình BERT đã được huấn luyện trước.
# BertForSequenceClassification.from_pretrained: Khởi tạo mô hình phân loại BERT từ mô hình đã được huấn luyện trước
#với số lượng nhãn bằng num_labels.

# Create DataLoader
train_dataset = CustomDataset(train_df, tokenizer, max_length=64)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
# train_dataset: Tạo đối tượng CustomDataset từ train_df, tokenizer, và với độ dài tối đa thực tế.
# DataLoader: Tạo DataLoader từ train_dataset với batch_size=8 và shuffle=True để ngẫu nhiên hóa dữ liệu trong quá trình huấn luyện

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
# Training nhanh hơn khi có gpu

# Initialize optimizer
optimizer = AdamW(model.parameters(), lr=2e-5) # lr: learing rate

# Training loop
num_epochs = 50 # Số vòng lặp huấn luyện
for epoch in range(num_epochs):
    model.train() # Gọi model training
    total_loss = 0 # Tính tổng loss
    total_predictions = [] #  dự đoán
    true_labels = [] # Thực tế

    for batch in tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs}'): # tqdm: thanh tiến trình
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad() # zero gradient, đặt gradient = 0
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels) #  Chạy mô hình và lấy đầu ra
        loss = outputs.loss # loss đầu ra
        total_loss += loss.item() # tổng loss

        # Get predicted labels
        logits = outputs.logits # đầu ra của legits => logits thường là bước trước bước hàm
        # logit: Đây là các giá trị thực tế (không phải xác suất) đại diện cho mức độ tự tin của mô hình về mỗi lớp (Positive, Neutral, Negative).
        predictions = torch.argmax(logits, dim=1).cpu().numpy() # Lấy dự đoán từ logits.
        total_predictions.extend(predictions) # Thêm dự đoán (prediction) vào danh sách
        true_labels.extend(labels.cpu().numpy()) # label thực tế

        loss.backward() # Tính gradient
        optimizer.step() # Cập nhật trọng số của mô hình

    average_loss = total_loss / len(train_loader) # Trung bình loss trên bộ training
    accuracy = accuracy_score(true_labels, total_predictions)
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {average_loss}, Accuracy: {accuracy}')

# Save the model
model.save_pretrained('model_no_1_1')
```

# **SVM**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn import metrics
import pandas as pd
df= pd.read_excel('/content/drive/MyDrive/AI/dataKFC_Final.xlsx')

# Đọc dữ liệu từ tệp Excel

# Kiểm tra tên các cột
#print(df.columns)

# Giả sử cột chứa văn bản là 'Review' và cột chứa nhãn là 'Sentiment'
texts = df['Review']  # Cột chứa văn bản
labels = df['Sentiment']  # Cột chứa nhãn

# Split dữ liệu thành tập huấn luyện và tập kiểm tra
train_texts, valid_texts, train_y, valid_y = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Tạo vector hóa TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Giới hạn số lượng từ thành 5000 từ
xtrain_tfidf = tfidf_vectorizer.fit_transform(train_texts)
xvalid_tfidf = tfidf_vectorizer.transform(valid_texts)

# Tối ưu hóa mô hình SVM bằng GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100],
              'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}
svm_classifier = SVC()
grid_search = GridSearchCV(estimator=svm_classifier, param_grid=param_grid, scoring='accuracy', cv=5)
grid_search.fit(xtrain_tfidf, train_y)

# In ra các tham số tốt nhất và độ chính xác tương ứng
print("Best parameters found:", grid_search.best_params_)
print("Best accuracy found:", grid_search.best_score_)

# Sử dụng các tham số tốt nhất để huấn luyện mô hình SVM
best_svm_classifier = SVC(**grid_search.best_params_)
best_svm_classifier.fit(xtrain_tfidf, train_y)

# Dự đoán nhãn cho tập kiểm tra và tính độ chính xác
predictions = best_svm_classifier.predict(xvalid_tfidf)
accuracy = metrics.accuracy_score(predictions, valid_y)
print("Accuracy with SVM:", accuracy)
```

```python
def translate_and_append(item):
    try:
        translation = translator.translate(item, dest="vi")
        trans.append(translation.text)
        language.append("vi")
    except Exception as e:
        # Retry translation with a sleep (avoid API quota limitations)
        time.sleep(1)
        try:
            translation = translator.translate(item, dest="vi")
            trans.append(translation.text)
            language.append("vi")
        except Exception as e:
            # Try translating to English
            try:
                translation = translator.translate(item, dest="en")
                trans.append(translation.text)
                language.append("en")
            except Exception as e:
                # Use original text as fallback and mark as Portuguese
                trans.append(item)
                language.append("pt")
                print(f"Error translating: {e}")

trans = []
language = []

# Loop through each item in the 'text' column and translate it
for item in df['text']:
    translate_and_append(item)
```

```python
# IMPORT TH∆Ø VI·ªÜN
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, Dataset
from torch import nn
from tqdm import tqdm
import torch
from sklearn.metrics import accuracy_score

# Map sentiment strings to integers
sentiment_mapping = {
    'Positive': 2,
    'Neutral': 1,
    'Negative': 0
}

# ƒê·ªçc d·ªØ li·ªáu Dataframe
df = pd.read_excel('/content/drive/MyDrive/AI/dataKFC_Final.xlsx')
df['Sentiment'] = df['Sentiment'].map(sentiment_mapping)

# Split the data into train and test sets Chia d·ªØ li·ªáu TRAINING:TESTING = 80:20
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Create a PyTorch Dataset ƒê∆∞a d·ªØ li·ªáu v·ªÅ d·∫°ng tensor
class CustomDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        text = self.dataframe.iloc[idx]['Review']
        label = torch.tensor(self.dataframe.iloc[idx]['Sentiment'], dtype=torch.long)
        encoding = self.tokenizer(text, truncation=True, padding='max_length', max_length=self.max_length, return_tensors='pt')
        return {'input_ids': encoding['input_ids'].flatten(), 'attention_mask': encoding['attention_mask'].flatten(), 'labels': label}
# CustomDataset: L·ªõp Dataset t√πy ch·ªânh c·ªßa PyTorch ƒë·ªÉ x·ª≠ l√Ω d·ªØ li·ªáu vƒÉn b·∫£n v√† chuy·ªÉn ƒë·ªïi n√≥ th√†nh tensors.
# __init__: Kh·ªüi t·∫°o l·ªõp v·ªõi dataframe, tokenizer v√† max_length.
# __len__: Tr·∫£ v·ªÅ s·ªë l∆∞·ª£ng m·∫´u trong dataframe.
# __getitem__: L·∫•y m·ªôt m·∫´u t·ª´ dataframe t·∫°i ch·ªâ s·ªë idx, chuy·ªÉn ƒë·ªïi vƒÉn b·∫£n th√†nh tensor v√† tr·∫£ v·ªÅ dictionary ch·ª©a input_ids,
#attention_mask v√† label.

# Initialize the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
num_labels = len(df['Sentiment'])
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)
# BertTokenizer.from_pretrained: Kh·ªüi t·∫°o tokenizer t·ª´ m√¥ h√¨nh BERT ƒë√£ ƒë∆∞·ª£c hu·∫•n luy·ªán tr∆∞·ªõc.
# BertForSequenceClassification.from_pretrained: Kh·ªüi t·∫°o m√¥ h√¨nh ph√¢n lo·∫°i BERT t·ª´ m√¥ h√¨nh ƒë√£ ƒë∆∞·ª£c hu·∫•n luy·ªán tr∆∞·ªõc
#v·ªõi s·ªë l∆∞·ª£ng nh√£n b·∫±ng num_labels.

# Create DataLoader
train_dataset = CustomDataset(train_df, tokenizer, max_length=64)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
# train_dataset: T·∫°o ƒë·ªëi t∆∞·ª£ng CustomDataset t·ª´ train_df, tokenizer, v√† v·ªõi ƒë·ªô d√†i t·ªëi ƒëa th·ª±c t·∫ø.
# DataLoader: T·∫°o DataLoader t·ª´ train_dataset v·ªõi batch_size=8 v√† shuffle=True ƒë·ªÉ ng·∫´u nhi√™n h√≥a d·ªØ li·ªáu trong qu√° tr√¨nh hu·∫•n luy·ªán

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
# Training nhanh h∆°n khi c√≥ gpu

# Initialize optimizer
optimizer = AdamW(model.parameters(), lr=2e-5) # lr: learing rate

# Training loop
num_epochs = 50 # S·ªë v√≤ng l·∫∑p hu·∫•n luy·ªán
for epoch in range(num_epochs):
    model.train() # G·ªçi model training
    total_loss = 0 # T√≠nh t·ªïng loss
    total_predictions = [] #  d·ª± ƒëo√°n
    true_labels = [] # Th·ª±c t·∫ø

    for batch in tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs}'): # tqdm: thanh ti·∫øn tr√¨nh
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad() # zero gradient, ƒë·∫∑t gradient = 0
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels) #  Ch·∫°y m√¥ h√¨nh v√† l·∫•y ƒë·∫ßu ra
        loss = outputs.loss # loss ƒë·∫ßu ra
        total_loss += loss.item() # t·ªïng loss

        # Get predicted labels
        logits = outputs.logits # ƒë·∫ßu ra c·ªßa legits => logits th∆∞·ªùng l√† b∆∞·ªõc tr∆∞·ªõc b∆∞·ªõc h√†m
        # logit: ƒê√¢y l√† c√°c gi√° tr·ªã th·ª±c t·∫ø (kh√¥ng ph·∫£i x√°c su·∫•t) ƒë·∫°i di·ªán cho m·ª©c ƒë·ªô t·ª± tin c·ªßa m√¥ h√¨nh v·ªÅ m·ªói l·ªõp (Positive, Neutral, Negative).
        predictions = torch.argmax(logits, dim=1).cpu().numpy() # L·∫•y d·ª± ƒëo√°n t·ª´ logits.
        total_predictions.extend(predictions) # Th√™m d·ª± ƒëo√°n (prediction) v√†o danh s√°ch
        true_labels.extend(labels.cpu().numpy()) # label th·ª±c t·∫ø

        loss.backward() # T√≠nh gradient
        optimizer.step() # C·∫≠p nh·∫≠t tr·ªçng s·ªë c·ªßa m√¥ h√¨nh

    average_loss = total_loss / len(train_loader) # Trung b√¨nh loss tr√™n b·ªô training
    accuracy = accuracy_score(true_labels, total_predictions)
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {average_loss}, Accuracy: {accuracy}')

# Save the model
model.save_pretrained('model_no_1_1')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn import metrics
import pandas as pd
df= pd.read_excel('/content/drive/MyDrive/AI/dataKFC_Final.xlsx')

# ƒê·ªçc d·ªØ li·ªáu t·ª´ t·ªáp Excel

# Ki·ªÉm tra t√™n c√°c c·ªôt
#print(df.columns)

# Gi·∫£ s·ª≠ c·ªôt ch·ª©a vƒÉn b·∫£n l√† 'Review' v√† c·ªôt ch·ª©a nh√£n l√† 'Sentiment'
texts = df['Review']  # C·ªôt ch·ª©a vƒÉn b·∫£n
labels = df['Sentiment']  # C·ªôt ch·ª©a nh√£n

# Split d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† t·∫≠p ki·ªÉm tra
train_texts, valid_texts, train_y, valid_y = train_test_split(texts, labels, test_size=0.2, random_state=42)

# T·∫°o vector h√≥a TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Gi·ªõi h·∫°n s·ªë l∆∞·ª£ng t·ª´ th√†nh 5000 t·ª´
xtrain_tfidf = tfidf_vectorizer.fit_transform(train_texts)
xvalid_tfidf = tfidf_vectorizer.transform(valid_texts)

# T·ªëi ∆∞u h√≥a m√¥ h√¨nh SVM b·∫±ng GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100],
              'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}
svm_classifier = SVC()
grid_search = GridSearchCV(estimator=svm_classifier, param_grid=param_grid, scoring='accuracy', cv=5)
grid_search.fit(xtrain_tfidf, train_y)

# In ra c√°c tham s·ªë t·ªët nh·∫•t v√† ƒë·ªô ch√≠nh x√°c t∆∞∆°ng ·ª©ng
print("Best parameters found:", grid_search.best_params_)
print("Best accuracy found:", grid_search.best_score_)

# S·ª≠ d·ª•ng c√°c tham s·ªë t·ªët nh·∫•t ƒë·ªÉ hu·∫•n luy·ªán m√¥ h√¨nh SVM
best_svm_classifier = SVC(**grid_search.best_params_)
best_svm_classifier.fit(xtrain_tfidf, train_y)

# D·ª± ƒëo√°n nh√£n cho t·∫≠p ki·ªÉm tra v√† t√≠nh ƒë·ªô ch√≠nh x√°c
predictions = best_svm_classifier.predict(xvalid_tfidf)
accuracy = metrics.accuracy_score(predictions, valid_y)
print("Accuracy with SVM:", accuracy)
```
